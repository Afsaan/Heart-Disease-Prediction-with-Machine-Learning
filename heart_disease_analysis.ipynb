```python
# importing all the important libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
import warnings
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
warnings.filterwarnings('ignore')
```

```python
#KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Random forest classifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# reading the dataset
data_frame = pd.read_csv('heart.csv')
```

# Feature Selection

```python
# creating a co-relation graph between all the varibale
corr_mat = data_frame.corr()
top_corr_features = corr_mat.index
plt.figure(figsize=(16,16))

#plot the heat map
heat_map =sns.heatmap(data_frame[top_corr_features].corr(),annot=True,cmap="RdYlGn")
plt.show()

#saving the figure as png file
heat_map.figure.savefig('corr_mat.png')
```

```python
# check if the target classes are of approximately equal size or not
sns.set_style('whitegrid')
sns_plot = sns.countplot(x='target',data=data_frame,palette='RdBu_r')
plt.show()

#saving the figure as png file
sns_plot.figure.savefig('target_count.png')
```

# Data Preprocessing

need to convert some categorical variables into dummy variables and scale all the values before training the Machine Learning models

```python
# defining the standarScaler
standardScaler = StandardScaler()

columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
data_frame[columns_to_scale] = standardScaler.fit_transform(data_frame[columns_to_scale])
data_frame.head()
```

```python
# splitting the independent and dependent variable

X = data_frame.drop(['target'], axis = 1)
y = data_frame['target']
```

# KNeighborsClassifier

```python
knn_scores = []
for k in range(1,21):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    score=cross_val_score(knn_classifier,X,y,cv=10)
    knn_scores.append(score.mean())
```

```python
plt.figure(figsize=(20, 10))
plt.plot([k for k in range(1, 21)], knn_scores, color = 'red')
for i in range(1,21):
    plt.text(i, knn_scores[i-1], (i, knn_scores[i-1]))
plt.xticks([i for i in range(1, 21)])
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Scores')
plt.title('K Neighbors Classifier scores for different K values')

#saving the figure
plt.savefig('kNN.png')

plt.show()
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors = 18)
score=cross_val_score(knn_classifier,X,y,cv=10)
score.mean()
```

# Random forest

```python
randomforest_classifier= RandomForestClassifier(n_estimators=10)
score=cross_val_score(randomforest_classifier,X,y,cv=10)
score.mean()
```
